# Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. Matplotlib is a multi-platform data visualization library built on NumPy arrays. One of the greatest benefits of visualization is that it allows us visual access to huge amounts of data in easily digestible visuals. Matplotlib consists of several plots like line, bar, scatter, histogram etc.

    1. Line Plot: Line plots are drawn by joining straight lines connecting data points where the x-axis and y-axis values            intersect.
    2. Bar Plot : The bar plots are vertical/horizontal rectangular graphs that show data comparison where you can gauge the          changes over a period represented in another axis (mostly the X-axis). Each bar can store the value of one or multiple          data divided in a ratio. 
    3. Scatter Plot: We can implement the scatter (previously called XY) plots while comparing various data variables to              determine the connection between dependent and independent variables. The data gets expressed as a collection of points        clustered together meaningfully.
    4. Pie Plot:  A pie plot is a circular graph where the data get represented within that components/segments or slices of          pie. Data analysts use them while representing the percentage or proportional data in which each pie slice represents an        item or data classification. 
    5. Histogram Plot: We can use a histogram plot when the data remains distributed, whereas we can use a bar graph to compare        two entities. Both histogram and bar plot look alike but are used in different scenarios.

# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

Scatter Plot: We can implement the scatter (previously called XY) plots while comparing various data variables to              determine the connection between dependent and independent variables. The data gets expressed as a collection of points        clustered together meaningfully.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.scatter(x,y)
plt.xlabel("This is x axis")
plt.ylabel("This is y axis")
plt.title("Random no. plotting")
```

# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

subplot() is used to draw multiple plots in one figure. The subplot() function takes three arguments that describes the layout of the figure.

The layout is organized in rows and columns, which are represented by the first and second argument.

The third argument represents the index of the current plot.

plt.subplot(1, 2, 1)
#the figure has 1 row, 2 columns, and this plot is the first plot.

```python


# Plot line 1

x = np.array([0, 1, 2, 3, 4, 5])  
y = np.array([0, 100, 200, 300, 400, 500])

plt.subplot(2,2,1)
plt.plot(x,y)
plt.title("Line 1")

# Plot Line 2

x = np.array([0, 1, 2, 3, 4, 5])  
y = np.array([50, 20, 40, 20, 60, 70])

plt.subplot(2,2,2)
plt.plot(x,y)
plt.title("Line 2")

# Plot Line 3
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([10, 20, 30, 40, 50, 60])

plt.subplot(2,2,3)
plt.plot(x,y)
plt.title("Line 3")

# Plot Line 4
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2,2,4)
plt.plot(x,y)
plt.title("Line 4")

plt.show()
```

# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

Bar Plot : The bar plots are vertical/horizontal rectangular graphs that show data comparison where you can gauge the          changes over a period represented in another axis (mostly the X-axis). Each bar can store the value of one or multiple          data divided in a ratio.

1. Visulize trends over time
2. Show distribution of data
3. Display and compare the size or frequency


```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Vertical bar chart
plt.bar(company,profit)
plt.xlabel("Company")
plt.ylabel("Profit")
plt.title("Profit by IT companies")
plt.show()
```

```python
# Horizontal bar chart
plt.barh(company,profit)
plt.xlabel("Profit")
plt.ylabel("Company")
plt.title("Profit by IT companies")
plt.show()
```

# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

A Box Plot is also known as Whisker plot is created to display the summary of the set of data values having properties like minimum, first quartile, median, third quartile and maximum. In the box plot, a box is created from the first quartile to the third quartile, a vertical line is also there which goes through the box at the median. Here x-axis denotes the data to be plotted while the y-axis shows the frequency distribution.

To display the summary of the set of data values having properties like minimum, first quartile, median, third quartile and maximum

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
box=[box1,box2]

# Title
plt.title("Box plot of Box1 and Box2")

plt.boxplot(box, patch_artist=True,labels=['Box1','Box2'])


plt.show()


 
```

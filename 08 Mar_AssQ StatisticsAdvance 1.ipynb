# Q1. What is the Probability density function?

The Probability Density Function(PDF) defines the probability function representing the density of a continuous random variable lying between a specific range of values. In other words, the probability density function produces the likelihood of values of the continuous random variable. However, this function is stated in many other sources as the function over a broad set of values. Often it is referred to as cumulative distribution function or sometimes as probability mass function(PMF). However, the actual truth is PDF (probability density function ) is defined for continuous random variables, whereas PMF (probability mass function) is defined for discrete random variables.

In the case of a continuous random variable, the probability taken by X on some given value x is always 0. In this case, if we find P(X = x), it does not work. Instead of this, we must calculate the probability of X lying in an interval (a, b). Now, we have to figure it for P(a< X< b), and we can calculate this using the formula of PDF. # Poisson Distribution(PMF)

 Poisson Distribution is applicable in situations where events occur at random points of time and space wherein our interest lies only in the number of occurrences of the event.
A distribution is called a Poisson distribution when the following assumptions are valid:

1. Any successful event should not influence the outcome of another successful event.
2. The probability of success over a short interval must equal its probability over a longer interval.
3. The probability of success in an interval approaches zero as the interval becomes smaller.

if any distribution validates the above assumptions, then it is a Poisson distribution. Some notations used in Poisson distribution are:

λ is the rate at which an event occurs,
t is the length of a time interval,
And X is the number of events in that time interval.
Here, X is called a Poisson Random Variable, and the probability distribution of X is called Poisson distribution.

# Q3. Write a Python function to calculate the probability density function of a normal distribution with given mean and standard deviation at a given point.

```python
from scipy.stats import norm
import numpy as np
  
data_start = -5
data_end = 5
data_points = 11
data = np.linspace(data_start, data_end, data_points)
print(data)  

# Calculating mean and Std. deviation.
mean = np.mean(data)
std = np.std(data)

print("Mean :",mean)
print("Std. Deviation :",std)
  
# Calculating probability density function
probability_pdf = norm.pdf(3, loc=mean, scale=std)
print("PDF :",probability_pdf)
```

# Q5. Generate a random sample of size 1000 from a binomial distribution with probability of success 0.4 and plot a histogram of the results using matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt

sample_size=1000
P_success = 0.4

# Generating random sample
x = np.random.binomial(sample_size, P_success, size=1000)

# Graph plotting
plt.figure(figsize=(15,10))
plt.hist(x, bins=30, density=True, alpha=.8)
plt.xlabel('No. of success')
plt.ylabel('Density of probability')
plt.title('Probability of success us Binomial Dist.')
plt.show
```

# Q7. How Binomial distribution different from Poisson distribution?

The binomial distribution is one, whose possible number of outcomes are two, i.e. success or failure. On the other hand, there is no limit of possible outcomes in Poisson distribution

Binomial distribution is one in which the probability of repeated number of trials are studied. Poisson Distribution gives the count of independent events occur randomly with a given period of time.

Binomial distribution is Biparametric where as Poisson distribution is Uniparametric.

Binomial distribution is fixed where as poisson distribution is infinite.

Binomial distribution is constant probability where as poisson distribution has Infinitesimal chance of success.

In binomial distribution Mean > Variance where as in poisson mean = variance


# Q8. Generate a random sample of size 1000 from a Poisson distribution with mean 5 and calculate the sample mean and variance.

```python
import numpy as np

# given mean = 5
mean=5

# random sample
sample_data = np.random.poisson(mean, size=1000)

# sample mean and variance

sample_mean = np.mean(sample_data)
sample_variance = np.var(sample_data)

print(sample_mean)
print(sample_variance)
                     
```

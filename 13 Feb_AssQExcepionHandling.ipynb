# Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

```python
# The lookupError exceptions are raised when an index or key is not found for a sequence and dictionary respectively. LookupError
# exception class handle both IndexError and KeyError exception classes.
```

```python
x = [1, 2, 3, 4]
try:
    print(x[10])
except LookupError as e:
    print(f"{e}, {e.__class__}")
```
